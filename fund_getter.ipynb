## 获取美股基金

```python
import akshare as ak
import pandas as pd
```

```python
# 获取所有基金
all_fund = ak.fund_name_em()

# 搜索包含“纳指”或“标普”的行
keywords = ["纳指", "标普", "美国", "纳斯达克"]

# 排除的关键词
exclude_keywords = ["医疗", "生物", "石油", "油气", "中国", "港股", "房地产", "道琼斯"]


pattern = "|".join(keywords)  # 构造正则表达式 "纳指|标普"
exclude_pattern = "|".join(exclude_keywords)

choose_fund = all_fund[
    all_fund["基金简称"].str.contains(pattern, na=False) &
    ~all_fund["基金简称"].str.contains(exclude_pattern, na=False)
]

print(f"匹配到的基金数量：{choose_fund.shape}")
choose_fund
```

## 基金申购状态

```python
fund_purchase_state = ak.fund_purchase_em() # 基金申购状态
choose_fund.groupby(by=["基金代码"], as_index=False)
merged_df = pd.merge(choose_fund, fund_purchase_state, on="基金代码", how="inner", suffixes=("", "_r"))  # 合并选择的基金和申购状态
merged_df
```

```python
# 删除重复列
columns_to_drop = [col for col in merged_df.columns if col.endswith("_r")]
merged_df = merged_df.drop(columns=columns_to_drop)
merged_df = merged_df.drop(columns=["序号", "拼音全称"])
merged_df
```

```python
# merged_df["购买起点"] = merged_df["购买起点"].apply(lambda x: f"{int(x):d}")
# merged_df["日累计限定金额"] = merged_df["日累计限定金额"].apply(lambda x: f"{int(x):d}")
can_buy = merged_df[(merged_df["日累计限定金额"] > 10) & (merged_df["申购状态"] != "暂停申购")]
```

```python
can_buy
```

```python
can_buy["购买起点"] = can_buy["购买起点"].apply(lambda x: f"{int(x):d}")
can_buy["日累计限定金额"] = can_buy["日累计限定金额"].apply(lambda x: f"{int(x):d}")
print(can_buy.shape)
print(can_buy.to_string())
```
